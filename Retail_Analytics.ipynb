```python
#  Question 4
#  Apply this model to the same data that the model was trained on and assess the prediction accuracy.
import pandas as pd
import statsmodels.api as sm
df = df_Activity
dummy_genders = pd.get_dummies(df['gender'], prefix = 'gender')
#print(dummy_genders)
dummy_genders.head()
dummy_metro = pd.get_dummies(df['metropolitan_area'], prefix = 'metro_area')
#print(dummy_metro)
dummy_device = pd.get_dummies(df['device_type'], prefix = 'device')
#print(dummy_device)
cols_to_keep = ['active', 'age']
activity_data = df[cols_to_keep].join(dummy_genders.loc[:, 'gender_M':])
#print(activity_data)
activity_data = activity_data.join(dummy_metro.loc[:, 'metro_area_Birmingham':])
#print(activity_data)
activity_data = activity_data.join(dummy_device.loc[:, 'device_Mobile':])
#print(activity_data)
activity_data = sm.add_constant(activity_data, prepend=False)
#print(activity_data)
explanatory_cols = activity_data.columns[1:]
print(explanatory_cols)
full_logit_model = sm.GLM(activity_data['active'],
activity_data[explanatory_cols], family=sm.families.Binomial())
result = full_logit_model.fit()
print (result)
result.summary()
```

```python
from sklearn import metrics
def confusion(y, X):
    prediction = []
    for number in result.predict(X).tolist():
        if number >= .5:
            prediction.append(1)
        else:
            prediction.append(0)

    print(metrics.confusion_matrix(y,prediction))
    holder = metrics.confusion_matrix(y,prediction)
    TP = holder[0][0]
    TN = holder[1][1]
    FP = holder[1][0]
    FN = holder[0][1]
    
    print("Accuracy: ", (TP + TN) / (TP + TP + FN + FP))
    print("Precision: ", (TN) / (TN + FN))
    print("fhf")
```

```python
confusion(activity_data['active'], activity_data[explanatory_cols])
```
